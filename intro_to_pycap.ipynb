# Using the REDCap API in Python Applications

[REDCap](http://project-redcap.org) is an advanced web-based application for securely storing and retrieving any type of information. In simple terms, it can be thought of a web-based spreadsheet, though it is much more than that. It provides an **Application Programming Interface** which means external software can programmatically download and upload data into REDCap Projects.

Becuase the API is based on simple HTTP requests, any programming langauge with a HTTP library can use the REDCap API. I predominantly write scripts and applications in Python. To make it easier to use the REDCap API from within python scripts and applications, I wrote [`PyCap`](http://sburns.github.io/PyCap).

This notebook is an easy tutorial about using the `PyCap` package. It assumes a Mac OS X or Linux environment, though there's no reason any of this wouldn't work on Windows. It assumes working knowledge of the shell and the python language.

First, we must install the package. In a shell:

`$ pip install PyCap`

If you don't have `pip` installed, this will work (you really should though, `easy_install` is considered deprecated by much of the python community):

`$ easy_install PyCap`

You may notice another package, `requests`, is installed as well. `requests` is a fantastic 3rd-party HTTP library for python and `PyCap` uses it behind the scenes for all API calls to REDCap.

With installation out of the way, let's start writing python. We'll begin with importing the package. The two main classes your scripts and applications should use are the `Project` class and `RedcapError` exception.

```python
from redcap import Project, RedcapError
```

## Connecting to REDCap Projects

To use the REDCap API (regardless of whether you're using `PyCap`, you must know two things:

*   The API URL for your site's REDCap installation. For Vanderbilt, this url is `https://redcap.vanderbilt.edu/api/`.
*   The API token for your Project. A token is generated by the REDCap administrators and connects your user account to a particular REDCap Project. Therefore, if you have API access to many Projects, you will have many tokens to manage.
